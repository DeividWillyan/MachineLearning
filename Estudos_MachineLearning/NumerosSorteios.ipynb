```python
df = df.rename(index=str, columns={"1ª Dezena": "dezena1", 
                                   "2ª Dezena": "dezena2", 
                                   "3ª Dezena": "dezena3", 
                                   "4ª Dezena": "dezena4", 
                                   "5ª Dezena": "dezena5", 
                                   "6ª Dezena": "dezena6"})
df.head()
```

```python
import seaborn as sns

fig = plt.figure(figsize=(20,20))
fig.subplots_adjust(hspace=0.1, wspace=0.1)
for i in range(1,7):    
    str_dezena = 'dezena' + str(i)    
    n = df.groupby(str_dezena)[str_dezena].count().sort_values(ascending=False)
    n = n[n > 25]

    ax = fig.add_subplot(2, 3, i)
    descending_order = n.index
    ax = fig.add_subplot(sns.countplot(y=df[str_dezena],order=descending_order))


    plt.yticks(fontsize=9)
    plt.xticks(fontsize=9)
    ax.plot()
```
