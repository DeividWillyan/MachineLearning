This csv contains information about all of the battles in game of thrones.

<b>---Colunas---</b>

<b>name</b> - The name of the battle.<br>
<b>year</b> - The year of the battle.<br>
<b>battle_number</b> - A unique ID number for the battle.<br>
<b>attacker_king</b> - The attacker's king. A slash indicators that the king charges over the course of the war. For example, "Joffrey/Tommen Baratheon" is coded as such because one king follows the other in the Iron Throne.<br>
<b>defender_king</b> - The defender's king.<br>
<b>attacker_1</b> - Major house attacking.<br>
<b>attacker_2</b> - Major house attacking.<br>
<b>attacker_3</b> - Major house attacking.<br>
<b>attacker_4</b> - Major house attacking.<br>
<b>defender_1</b> - Major house defending.<br>
<b>defender_2</b> - Major house defending.<br>
<b>defender_3</b> - Major house defending.<br>
<b>defender_4</b> - Major house defending.<br>
<b>attacker_outcome</b> - The outcome from the perspective of the attacker. Categories: win, loss, draw.<br>
<b>battle_type</b> - A classification of the battle's primary type. Categories: pitched_battle: Armies meet in a location and fight. This is also the baseline category. ambush: A battle where stealth or subterfuge was the primary means of attack. siege: A prolonged of a fortied position. razing: An attack against an undefended position<br>
<b>major_death</b> - If there was a death of a major figure during the battle.<br>
<b>major_capture</b> - If there was the capture of the major figure during the battle.<br>
<b>attacker_size</b> - The size of the attacker's force. No distinction is made between the types of soldiers such as cavalry and footmen.<br>
<b>defender_size</b> - The size of the defenders's force. No distinction is made between the types of soldiers such as cavalry and footmen.<br>
<b>attacker_commander</b> - Major commanders of the attackers. Commander's names are included without honorific titles and commanders are separated by commas.<br>
<b>defender_commander</b> - Major commanders of the defender. Commander's names are included without honorific titles and commanders are separated by commas.<br>
<b>summer</b> - Was it summer?<br>
<b>location</b> - The location of the battle.<br>
<b>region</b> - The region where the battle takes place. Categories: Beyond the Wall, The North, The Iron Islands, The Riverlands, The Vale of Arryn, The Westerlands, The Crownlands, The Reach, The Stormlands, Dorne.<br>
<b>note</b> - Coding notes regarding individual observations.<br>

```python
#Bibliotecas
import pandas as pd
```

```python
#Importação dos dados pelo pandas.

df = pd.read_csv('battles.csv')
df['attacker_outcome'] = df.attacker_outcome.map({'win':0, 'loss':1})
df = df.fillna(0)

df.head(3)
```

```python
# Separando os dados

resultado = df.attacker_outcome
parametros = df.drop(['attacker_outcome'], axis=1)
#resultado = resultado.values.reshape(-1,1)

print('O Conjunto de parametros tem %s linhas e %s colunas.' % (parametros.shape))
```

```python
# Verificando a correlação entre as variaveis.

#parametros.corr()

#Removendo variaveis sem relação
parametros = parametros.drop(['defender_3', 'defender_4', 'note'], axis=1)

print(parametros.corr())
```

```python
# tratando as variaveis tipo string

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in parametros.columns:
    if parametros[column_name].dtype == object:
        parametros[column_name] = le.fit_transform(parametros[column_name])
    else:
        pass
    

print(parametros.describe)
```

```python
# Utilizarei o classificador RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(parametros, resultado, test_size=0.33, random_state=42)

clf = RandomForestClassifier(n_estimators = 100, min_samples_leaf = 10)
clf.fit(X_train, y_train)

predicoes = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, predicoes)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, predicoes))

for i,j in zip(y_test, predicoes):
    if (i == 0):
        i = 'win'
    else:
        i = 'loss'
    if (j == 0):
        j = 'win'
    else:
        j = 'loss'
    print('Era pra ser: %s mas previu %s' % (i, j))
```
